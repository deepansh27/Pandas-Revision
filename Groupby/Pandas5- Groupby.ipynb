```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
df = pd.read_csv('weather_by_cities.csv')
```

```python
df
```

```python
g = df.groupby('city')
g
```

We can see the pictorical representation of the groupby as follows:


```python
# printing the groups with its valaues
for group, grp_df in g:
    print(group)
    print(grp_df)
```

```python
# to access specific group
g.get_group('mumbai') 

# Note: the above is similar to 
# SELECT * from city_data GROUP BY city
```

```python
g.max()
```

```python
g.mean()
```
